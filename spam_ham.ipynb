```python
import pandas as pd
messages= pd.read_csv('D:\spam_ham\SMSSpamCollection', sep='\t', names=["labels","message"])
```

```python
messages.head()
```

```python
messages.shape
```

```python
import re
import nltk 
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
x=  WordNetLemmatizer()
# Lemmatize a word
corpus=[]
for i in range(len(messages)):
    review=re.sub('[^a-zA-Z0-9]', ' ', messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[x.lemmatize(word, pos="v") for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=2500)
x=cv.fit_transform(corpus).toarray()
```

```python
y=pd.get_dummies(messages['labels'])
print(y)
y=y.iloc[:,1].values
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=25)
```

```python
from sklearn.naive_bayes import MultinomialNB
mb=MultinomialNB()
mb.fit(x_test,y_test)
```

```python
y_pred=mb.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_train,y_pred)
```

```python
print(score)
```
