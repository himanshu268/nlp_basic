```python
import pandas as pd
df=pd.read_csv('train.csv',encoding='unicode-escape')
```

```python
df.isnull().sum()
```

```python
data=df.reset_index(drop=True)
```

```python
data.isnull().sum()
```

```python
y = data['sentiment'].values.ravel()
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
z=WordNetLemmatizer()
```

```python
data['selected_text'][315]
# data.isnull().sum()
```

```python
data.shape
```

```python
corpus=[]
for i in range(len(data)):
    text = str(data['selected_text'][i])  # Convert to string, handling NaNs and other types
    review = re.sub('[^a-zA-Z0-9]', ' ', text)
    review = review.lower()
    review = review.split()
    # corpus.append(review)
    review=[z.lemmatize(word,pos='v')for word in review if not word in stopwords.words('english')]
    review=" ".join(review)
    corpus.append(review)
    # print(corpus)
```

```python
corpus
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(max_features=2500)
x=tf.fit_transform(corpus).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
mb=MultinomialNB()
mb.fit(x,y)
```
